```
Row Subset을 slicing, indexing, condition 중에서 진행한다고 했을 때 condition의 의미는?


"Condition"이란 일반적으로 특정 조건에 따라 데이터를 선택하는 것을 의미. 
pandas에서는 이를 "boolean indexing" 또는 "conditional indexing"이라고도 함.

예를 들어, DataFrame에서 'SepalLengthCm'이 5.0보다 큰 모든 행을 선택하려면 다음과 같이 할 수 있음:


df[df['SepalLengthCm'] > 5.0]

위 코드에서 df['SepalLengthCm'] > 5.0은 각 행에 대해 'SepalLengthCm' 값이 5.0보다 큰지 여부를 True/False로 반환하는 조건임. 
그리고 df[...]는 이 조건에 맞는 (즉, True인) 행만 선택.

또한 여러 조건을 결합하여 복잡한 쿼리를 만들 수도 있음:


df[(df['SepalLengthCm'] > 5.0) & (df['Species'] == 'Iris-setosa')]
위 코드는 'SepalLengthCm' 값이 5.0보다 크고, 동시에 'Species'가 'Iris-setosa'인 모든 행을 선택함.

따라서 "condition"은 데이터의 값을 검사하여 원하는 데이터만 추출하는 방법을 의미함.
```
