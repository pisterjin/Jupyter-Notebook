## make_blobs를 사용한 분류학습용 데이터 생성
- 모델에 정답을 알려주는 Classification 
- 그리고 모델에 정답을 알려주지 않는 클러스터링

````python
# 회귀에서의 문제/정답 => 수치 개념의 의미
# 분류에서의 문제/정답 => 수치 개념이 아닌 이름을 가리키는 것뿐임
# Classification =>  분류를 의미
````

````python
# 데이터 셋 준비
from sklearn.datasets import make_blobs     # blobs는 작은 구체' 또는 '덩어리'라는 뜻
import matplotlib.pyplot as plt
````

````python
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=0)
# n_samples=1000: 전체 데이터 포인트 개수가 1000개인 데이터셋을 생성하겠다는 것을 나타냄. 
# 예를 들어, 공간상에 1000개의 점이 무작위로 흩어져 있음을 의미

# centers는 클러스터 or 클래스를 의미하며 데이터 포인트들의 그룹으로, 서로 유사한 특성을 가진 데이터 포인트들이 모여있는 영역을 의미
# 데이터 포인트를 모아 놓은 3개의 클러스터를 생성하겠다는 것

# n_features=2: 이는 각 데이터 포인트가 2개의 속성(피처)을 가질 것임을 나타냄. 여기서 2개의 속성은 2차원 좌표 (x, y)를 의미.
````

````python
plt.scatter(X[:,0],X[:,1], s=3)  # 시스템에서 만들어준 데이터

# X[:,0]=> X 역할을 하는 데이터의 집합/ X[:,1]=> Y 역할을 하는 데이터의 집합
# X,Y는 한 점의 좌표이기 때문에, 이 점 한 개가 어느 그룹에 속해있는지 정답이 있어야 함
# centers가 3이므로 각 점은 그룹 0, 1, 2 중 하나인데 y가 정답이므로 0, 1, 2중 어떤 값인지 표기가 되어 나옴
# y가 0, 1, 2중 하나이므로 c=y라고 두면 y값마다 세 가지 색상 중 하나를 가지게 됨
plt.show()
````

````python
# (그림만 두고 볼 때) 위와 같은 그림은 정답을 안 주고 나누었을 때이며 비지도 학습이라고 하여 Classification라고 하지 않고 
# 그 군집을 클러스터링이라고 함
# 그리고 그 덩어리를 클라스터라고 함
````

````python
plt.scatter(X[:,0],X[:,1], c=y, s=3)  # 시스템에서 만들어준 데이터

# X[:,0]=> X 역할을 하는 데이터의 집합/ X[:,1]=> Y 역할을 하는 데이터의 집합
# X,Y는 한 점의 좌표이기 때문에, 이 점 한 개가 어느 그룹에 속해있는지 정답이 있어야 함
# centers가 3이므로 각 점은 그룹 0, 1, 2 중 하나인데 y가 정답이므로 0, 1, 2중 어떤 값인지 표기가 되어 나옴
# y가 0, 1, 2중 하나이므로 c=y라고 두면 y값마다 세 가지 색상 중 하나를 가지게 됨
plt.show()
````

````python
# 위 내용은 지도 학습에 해당되며 분류되어 있는 데이터가 0, 1, 2 중 어느 그룹에 속하는지를 학습시키는 것
# 점들의 위치를 알기 위해 학습를 하게 되고 데이터 위치는 주면서 분류의 기준을 만들라고 하는 것
# 학습을 통해 계수와 절편을 구하게 됨
````

````python
# 데이터 셋 준비
from sklearn.datasets import make_blobs     # blobs는 작은 구체' 또는 '덩어리'라는 뜻
import matplotlib.pyplot as plt
````

````python
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=0)
X.shape, y.shape
````

````python
X[:6],y[:6]   # [:5]로 하면 1 그룹이 안 나옴   X 6개와 y 6개를 보여달라는 의미

# 좌표 [-0.80723584,  2.66946295]가 2그룹이라는 의미
````

````python
plt.scatter(X[:,0],X[:,1], s=3)  # 시스템에서 만들어준 데이터
plt.show()   # 색을 없게 해도 답을 줬다면 클러스터가 아닌 Classification
````

````python
plt.scatter(X[:,0], X[:,1], c=y, s=3, )
plt.show()
````

````python
# 클래스에 따라서 직접 색상을 지정하는 예

col = {0:'r',1:'g',2:'b'}
# 리스트 컴프리헨션(List comprehension) 사용 
color = [col[c] for c in y]  # 점이 1000개니까 1000개가 색으로 나와야 함. col[c]가 키가 되고 y 0이면 r  
# (centers=3이므로 y는 0,1,2 중 하나이므로 숫자를 키로 주고 색상을 값으로 지정)
color[:5] # 값이 잘 나오는지 확인 위해 5개만 테스트해봄

# color = {col[c] for c in y} 이렇게 하면  set은 순서가 없는 자료형이라, 인덱싱을 통한 접근이 불가능해서
# color[:5] 로 출력시 오류 발생됨.
# 색상을 클래스에 매핑하려면 리스트나 넘파이 배열 등 순서가 있는 자료형을 사용해야 함
````

````python
# 리스트 컴프리헨션(List comprehension) => 리스트에 for와 if(선택적) 문을 사용하여 리스트를 생성하는 것

#  0부터 9까지의 숫자를 제곱한 리스트를 만드는 코드
squares = [x**2 for x in range(10)]
squares # squares는 [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]이라는 값들로 구성되는데, 이는 각 숫자(0부터 9까지)를 제곱한 값.
````

````python
# 리스트 컴프리헨션(List comprehension) => [표현식 for 항목 in iterable]

# 0부터 9까지의 숫자 중 짝수만 제곱하여 만든 리스트

even_squares = [x**2 for x in range(10) if x % 2 == 0]
even_squares  # even_squares는 [0, 4, 16, 36, 64]라는 값들로 구성되는데, 각 숫자는 짝수들(0, 2, 4, 6, 8)을 제곱한 것
````

````python
plt.scatter(X[:,0],X[:,1], c=color, s=3)  # 시스템에서 만들어준 데이터
plt.show()
````

````python
# legend를 사용하여 색상별 분류 클래스 표시의 예(레전드와 범례 주기 => 0 클래스는 보라색으로 되어 있다는 것 등을 보여줌)

fig, ax = plt.subplots()  # ax를 사용하기 위해 plt의 subplots() 함수 호출함. ax는 axis의 복수로 axes(또는 axises) '축들'을 의미
# fig는 바깥 라인 전체를 관리하는 객체
# ax는 fig(피규어) 안에 두 축 안의 데이터를 관리 객체
# 여기서는 전체 사이즈 관리할 필요 없어서 fig는 사용하지 않았음

scatter = ax.scatter(X[:,0],X[:,1], c=y, s=3, cmap='Reds')  # cmap='Reds' 없애면 시스템이 알아서 색상 생성
# 원칙은 위 내용처럼 해야 하나 plt.scatter만 해도 시스템 내부에서 ax.scatter를 호출해줌
# 그러나 차트를 좀 더 세밀하게 관리 위해서는 fig나 ax 객체를 따로 함수 호출해서 써야 함
# 예를 들면, ax 객체를 이용해 레전드(범례)를 만듬

# produce a legend with the unique colors from the scatter
# legend_elements() 레전드 요소를 가지고 레전드를 만듬
# ax.scatter(X[:,0],X[:,1], c=y, s=3, cmap='Reds') 코드로 이렇게 기본만 쓰면 
# 디폴트로 클래스로 볼 만한 게 'c=y'뿐이라서 컬러를 가지고 범례를 만듬
# y가 정답이니 y값을 가지고 범례를 만듬
legend1 = ax.legend(*scatter.legend_elements(),  
# 함수 선언시 파라미터 *은 가변인자-여러 개의 데이터-를 받아도 괜찮고 그것을 튜플로 패킹한다는 의미 
# 호출시 *는 클래스에서 원소를 꺼내 낱개 쓰기 위한 언팩 과정 => 여기서는 함수 호출이므로 언팩의 의미임
                      loc="upper right", title="Classes")
# loc="upper right", title="Classes" => 범례 위치는 오른쪽 상단, 제목은 분류라는 의미로 쓴 것
    
ax.add_artist(legend1)
plt.show()
````

# 함수 호출시의 '*'

- 함수 호출시의 '*'는 "unpacking" 연산자로 사용됨. 
- 주로 함수에 인자를 전달할 때 사용되며, 리스트나 튜플과 같은 시퀀스 타입의 데이터를 개별 요소로 분해하여 전달하는 역할을 함.
- 예를 들어, scatter_all.legend_elements()는 튜플을 반환. 이 튜플의 각 요소를 ax.legend() 함수에 개별적인 인자로 전달하려면 * 연산자를 사용해야 함.
- 간단한 예시입니다:

- 코드 그대로 띄어쓰기 포함해 쓰려면 역따옴표 세 개(앞 뒤로) 
```
def my_function(a, b, c):
    print(a, b, c)

my_tuple = (1, 2, 3)

my_function(*my_tuple)  # 출력: 1 2 3
```
- 위 코드에서 my_function(*my_tuple)는 my_function(1, 2, 3)과 동일하게 작동. 즉, 튜플의 각 요소가 함수의 각 매개변수에 순서대로 할당됨.
- scatter_all.legend_elements()는 레전드 (범례)를 만들기 위한 정보를 반환하는 함수. 
- 이 함수는 범례에 표시될 항목들과 그에 해당하는 레이블을 반환. 
- 그리고 *scatter_all.legend_elements()는 이 정보를 분리하여 함수 ax.legend()의 인자로 전달하는 역할을 함.
- *scatter_all.legend_elements()는 반환된 정보를 풀어서 함수에 전달하라는 의미이며, loc은 범례의 위치를 지정하고, title은 범례의 제목을 설정하는 것. 이렇게 하면 범례를 생성하고 설정할 수 있음.

# 함수 선언시의 '*'

- 함수 선언시에 *를 사용하는 것은 "가변 인자(variable-length arguments)"를 의미. 
- 이는 함수가 임의의 개수의 인자를 받을 수 있도록 해줌.

- 예를 들면

```
def my_function(*args):
    for arg in args:
        print(arg)
```        

- 이 함수는 몇 개의 인자가 전달되든 모두 처리할 수 있음  
- *args는 전달된 모든 위치 기반 인자들을 튜플로 묶음:

```
my_function(1, 2, 3)  
# 출력: 
# 1
# 2
# 3

my_function(4, 5)
# 출력: 
# 4
# 5
```

- 또한, ** 연산자는 키워드 기반의 가변 인자들을 딕셔너리로 묶음:
```
def my_function(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} = {value}")

my_function(a=1, b=2)
# 출력: 
# a = 1
# b = 2
```
- 따라서 함수 선언시에 사용하는 *와 **는 각각 위치 기반 및 키워드 기반 가변 인자들을 처리하는 방법임.
- 함수 선언 시에도 *을 사용하는 경우는 함수의 매개변수를 정의할 때이며 이것을 "가변 인자" 또는 "패킹 인자"라고 함.
- 가변 인자를 사용하면 함수가 임의 개수의 인자를 받을 수 있도록 할 수 있음. 
- 함수를 호출할 때 전달한 인자들은 튜플로 묶여서 해당 가변 인자에 전달됨.
- 예를 들어, 다음과 같이 함수를 정의할 수 있음:

```
def my_function(*args):
    for arg in args:
        print(arg)

my_function(1, 2, 3, 4, 5)
```
- 이 경우, my_function은 임의의 개수의 인자를 받을 수 있음. 
- 함수 내에서 args는 튜플로 처리되며, 호출할 때 전달한 인자들이 그 튜플에 포함됨.
- 함수를 유연하게 사용하고자 할 때 유용한 패턴 중 하나.

### ML 평가지표
*   <img src='https://cwisky.github.io/public/images/precision.jpg' width='40%' align='right'>    
- Accuracy(정확도) : 올바르게 예측된 데이터의 수를 전체 데이터의 수로 나눈 값  
- 태풍이 불 것으로 예보하는 시스템이 항상 태풍이 없다고 판단하면 정확도가 매우 높게 나타남  
- Precision(정밀도) : True으로 예측한 갯수 중에서 실제로 True인 경우의 수  
- Recall(재현율) : 실제 True인 데이터 중에서 모델이 True라고 인식한 수  
- F1 Score(조화평균) : precision, recall을 조합하여 하나의 통계치를 생성한 것    
    
- Accuracy : (TP + TN) / (TP + TN + FP + FN)  
- Precision : TP / (TP + FP)
- Recall : TP / (TP + FN)
- F1 Score : (2 * precision * recall) / (precision - recall)

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = df3.drop(columns=['target_yes'])  # 컬럼 수가 많을 때에는 각 컬럼 명시 방법보다 제거하려는 컬럼만 drop에 넣고 제외하여 저장.  
y = df3['target_yes']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# random_state=0는 안 써도 학습하는데 문제 없으나 여럿이 참여하는 학습이라면 학습횟수에 따라 결과도 동일해야 하므로 넣어줌
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
model = LogisticRegression(max_iter=5000)
````

````python
model.fit(X_train, y_train)
````
