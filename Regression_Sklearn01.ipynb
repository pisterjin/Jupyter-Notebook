# 머신 러닝 시작

```python
import numpy as np

def make_reg_data(x):    # w와 b 를 컴퓨터가 맞추는 게 머신러닝의 핵심 x가 문제, y가 정답이고 문제와 답을 줌 
    w = 12.345
    b = -3.456
    y = w*x + b
    return y
```

```python
# x값 생성
x = np.arange(1000)
x[:10],x[990:]
```

```python
# y값 생성
y = make_reg_data(x)
y[:10], y[990:]
```

```python
x[:5],y[:5]
```

```python
# 학습용 데이터는 1차원이 아닌 2차원으로 줘야 함
```

```python
X = x.reshape(-1,1)   
X.shape

# reshape 함수에서 -1은 '나머지 차원'을 의미, 이 경우에는 모든 요소의 수를 나타냄. 
# 따라서 -1이 들어간 위치의 차원 크기는 자동 계산되어 결정
```

```python
X[:5]
```

```python
# sklearn Regression   # 사이킷런(sklearn) => 선형회귀
# scikit-learn의 줄임말
# scikit은 science kit[과학 도구/과학 도구 모음]의 줄임말

# 학습용과 검증용 데이터는 따로 써야 함(학습에 참여하지 않은 데이터를 이용해 만들어진 모델을 검증해야 정확성이 인증되므로)
# 데이터로 직선의 방정식 추정(즉, 가중치와 bias[절편 알아내기])

from sklearn.linear_model import LinearRegression    # linear 선으로 된 Regression 회귀 
# => 선형 회귀 분석을 수행하는 데 사용되며, 데이터 간의 선형 관계를 찾아내고 예측 모델을 만드는 역할의 import

from sklearn.model_selection import train_test_split    # 주어진 데이터셋을 훈련용(train)과 검증용(test)으로 나눌 때 사용됨
```

```python
from matplotlib import rcParams         # 차트 이용 위함
rcParams['font.family'] = 'New Gulim'     # 한글 쓰고자 깨짐 방지용
rcParams['font.size'] = 10                # 글자 크기 조정용
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y,s=1)
plt.title('학습용 데이터 시각화')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)

# test_size=0.2는 전체 데이터의 20%를 테스트용으로 사용하겠단 의미 
# random_state=0 => 난수 생성기의 시드 값을 0으로 설정한다는 의미, 사이킷런의 여러 함수에서 재현 가능한 결과를 위해 사용
# => 난수 생성기의 시드 값으로 사용되며, 이 시드 값이 같으면 동일한 작업을 반복할 때마다 동일한 결과를 얻음.
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape  

# X_train.shape: 800행 1컬럼[2차원 배열] # y_train.shape: 800개의 원소[1차원 배열]
```

```python
# LinearRegression 클래스의 인스턴스를 생성하고 그 인스턴스를 model 변수에 할당
# 이 인스턴스는 선형 회귀 모델을 나타내며, 이 모델은 학습 데이터에 적합(fit)한 후에 예측(predict)을 수행하는 데 사용됨. 
# 즉, LinearRegression()은 아직 어떤 데이터도 받지 않았고
# 어떤 선형 관계도 학습하지 않은 초기 상태의 선형 회귀 모델임.


# 선형회귀 모델 생성
model = LinearRegression()   # 선형 회귀 모델 객체를 생성하고 이를 model 변수에 할당한다
model
```

```python
# model.coef_, model.intercept_   # 계수(w), 편향값(b)은 아직 생성되지 않는 상태라 오류남
```

```python
dir(model)
```

```python
dir(model.get_params)
```

```python
# 학습 : 입력된 데이터에 가중치(w)와 편향값(b)를 맞춰가는(fit) 과정이다   # 학습시키는 것
model.fit(X_train, y_train)   # 모델을 학습시킴. 즉, 주어진 X_train과 y_train을 이용해 선형 회귀 모델의 가중치를 최적화함
```

```python
# 학습과정에서 결정된 x의 계수와 절편 확인
model.coef_, model.intercept_    

# model.coef_: 독립 변수의 계수(또는 가중치)를 나타냄. 예측 결과에 얼마나 영향을 미치는지 보여줌
# => 선형 회귀 모델의 계수(coefficient) 또는 가중치(weight)를 의미. 피처의 중요도를 나타내는 지표

# model.intercept_:  y 절편 값을 나타냄. 즉, 모든 입력 특성이 0일 때 예측값은 intercept_가 됨
```

```python
# 결정 계수 => 회귀 모델의 예측 성능을 평가하는 지표
model.score(X_train, y_train)   # 1에 가까울수록 정확한 값임
```

```python
# 검증용 데이터를 사용하여 테스트
pred = model.predict(X_test)    # X_test는 학습하지 않은 데이터임
```

## 노이즈가 포함된 데이터의 경우

```python
noise = np.random.randn(1000)
```

```python
# 노이즈값을 데이터에 줬을 경우

import numpy as np

# x값을 받아서 y값을 생성하기 위한 함수,
# 가중치(w)와 편향값(b)를 무작위로 정해서 y값을 생성,
def make_reg_data_noise(x):    # w와 b 를 컴퓨터가 맞추는 게 머신러닝의 핵심 x가 문제, y가 정답이고 문제와 답을 줌 
    w = 12.345
    b = -3.456
    y = w*x + b
    noise = np.random.randn(1000)*1000
    return y+noise
```

```python
x = np.arange(1000)                    # x에 arange(1000) 이용해 1000개의 데이터를 줌
y = make_reg_data_noise(x)
x[:10], y[:10]
```

```python
from matplotlib import rcParams         # 차트 이용 위함
rcParams['font.family'] = 'New Gulim'     # 한글 쓰고자 깨짐 방지용
rcParams['font.size'] = 10                # 글자 크기 조정용
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, s=1)
plt.title('노이즈 데이터 시각화')
plt.show()    # 탐색적 데이터
```

```python
# reshape(-1, 1)은 NumPy 배열의 메서드로, 배열의 차원을 바꾸는 데 사용됩니다.
# -1은 나머지 차원을 자동으로 계산하도록 지시하고,
# (뒤의 1을 통해 한 개의 열을 만들고 나머지 행들을 마음대로 하라는 의미=> 1000개의 행 만들라고 한 것. '1000,1' 쓰는 것보다 간결) 
# 1은 새로운 배열의 열(column) 수를 1로 만듭니다. 
# 따라서 x.reshape(-1, 1)은 원래의 1차원 배열 x를 열이 하나인 2차원 배열로 바꿉니다. 
# 이렇게 하면 각 원소가 한 행에 하나씩 들어가며, 머신러닝 모델에 적합한 형태가 됩니다.
```

```python
X = x.reshape(-1, 1)   # 문제에 해당하는 것은 2차원으로 만들어야 함
# X.shape : (1000, 1)  => 1000개의 행과 한 개의 열인 2차원데이터. 
# x에 1000개의 값을 줬는데 컬럼을 1개로 두고 남은 원소들이 나머지 자리를 채우라는 의미라서 1000개의 행이 생기게 됨
X.shape, y.shape

# shape는 numpy 배열의 차원과 각 차원별 요소 수를 표현하는 속성.
# 2차원 배열에서 첫 번째 숫자는 행의 수를 나타내고 두 번째 숫자는 열의 수를 나타냄.
# 예시로, (3, 4)라면 이것은 3행 4열인 2차원 배열을 의미하며,
# (5,) 형태라면 이것은 길이가(요소수가)5인 일차원 벡터(배열)을 의미.

# y값을 결정하는 요소는 여러 개일 수 있어서 가중치는 원소가 하나여도 2차원으로 만들어줘야 함
```

```python
from sklearn.linear_model import LinearRegression  # 회귀
from sklearn.model_selection import train_test_split  # 학습용과 검증용 분리
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)  # random_state=0 는 random seed 같은 역할
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape   # train 훈련, test 검증 
```

```python
# 선형 회귀 모델 생성
model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)     # 학습시킴
```

```python
model.coef_, model.intercept_   # 언더스코어 필수   model.coef_ => 결정계수(가중치) / intercept => 절편
```

```python
# 결정계수  => 1에 가까울수록 데이터가 정확

model.score(X_train, y_train)   #  결정 계수를 반환. 회귀 모델이 얼마나 잘 맞았는지 평가하는 척도로, 1에 가까울수록 좋음
```

```python
# 검증용 데이터(X_test)를 사용하여 테스트

pred = model.predict(X_test)    # X_test 데이터에 대한 예측값(y값)을 생성
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y,s=1) # 훈련용과 테스트용 구분 전의 전체 데이터의 산점도 // s 는 size 의 약자로 제일 작은 사이즈가 1임
plt.title('노이즈 데이터 시각화')
plt.plot(X_test, pred, 'r-')        # 컴터가 추정한 값(데이터의 패턴을 발견해낸 것을 차트로) => 회귀선(패턴 일반화)
plt.show()                          # 탐색적 데이터
```

## sklearn의 선형회귀용 학습 데이터 생성 함수
- sklearn.datasets.make_regression

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=500, n_features=1, noise=20, random_state=2)  # noise=50 노이즈 50이면 노이즈 강한 정도에 속함
X.shape, y.shape
```

```python
# n_features는 생성되는 특징(또는 변수)의 수.
# make_regression 함수는 회귀 문제에 사용할 수 있는 임의의 데이터를 생성하는 함수. 
# 이 함수에서 n_samples은 생성할 샘플(데이터 포인트)의 수, n_features는 각 샘플에 대한 특징(독립 변수 또는 입력 변수)의 수를 지정
# 예를 들어, n_features=1이면 각 데이터 포인트에 대해 하나의 특징 값만 있음. 
# 이 경우, 데이터는 2차원 평면(x-y 평면) 위에 나타낼 수 있음. 
# 만약 n_features=2라면 각 데이터 포인트에 대해 두 개의 특징 값이 있으며, 
# 이 경우 데이터는 3차원 공간(x-y-z 공간) 위에 나타낼 수 있음.
# 따라서 위 코드에서 X,y = make_regression(n_samples=500, n_features=1, noise=50, random_state=2)은 
# 500개의 샘플을 생성하고 각 샘플은 하나의 특징 값을 가지며 노이즈가 추가된 회귀 문제용 데이터를 만드는 것.
```

```python
plt.scatter(X, y, s=1)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression    # linear 선으로 된 Regression 회귀 
# => 선형 회귀 분석을 수행하는 데 사용되며, 데이터 간의 선형 관계를 찾아내고 예측 모델을 만드는 역할의 import
from sklearn.model_selection import train_test_split 
from matplotlib import rcParams         # 차트 이용 위함
rcParams['font.family'] = 'New Gulim'     # 한글 쓰고자 깨짐 방지용
rcParams['font.size'] = 10       
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)
model.coef_, model.intercept_
```

```python
model.score(X_train, y_train)
```

```python
pred = model.predict(X_test) 
```

```python
# 상관계수가 가장 높은 것의 변수명과 상관계수출력하기

# 1)'charges' 행을 제외한 상관계수 행렬에서 최대 값을 찾기.
max_corr = matrix.loc[matrix.index != 'charges', 'charges'].max()

# matrix.loc[matrix.index != 'charges', 'charges'] 
# => 상관계수 행렬(matrix)의 인덱스 중에서 'charges'가 아닌 행들을 먼저 선택하고, 
# 그 행들에서 'charges'의 열의 값들을 가져오는 코드로  
# 'age', 'bmi', 'children' 행에 있는 'charges' 열의 값들만 선택되어 그 중 max를 구할 수 있음

# 2) 최대 상관계수에 해당하는 변수명(index)를 찾기.
max_corr_variable = matrix['charges'][matrix['charges'] == max_corr].index[0]

max_corr_variable, max_corr
```

```python
# index[0]을 사용한 이유는 matrix['charges'][matrix['charges'] == max_corr].index의 결과가 Index 객체를 반환하고, 
# 실제 변수명을 얻기 위해서는 리스트처럼 인덱싱을 사용하여 첫 번째 요소를 가져와야 하기 때문.

# matrix['charges'][matrix['charges'] == max_corr].index의 결과 예시는 다음과 같음.

# Index(['age'], dtype='object')
# 여기서는 ['age']만 포함된 Index 객체가 반환됨.
# 변수명을 단순하게 문자열로 얻기 위해서는 index[0]을 사용하여 첫 번째 요소를 불러와야 함.

# 따라서, index[0]이 없으면 결과로 변수명 문자열 대신 원하지 않는 Index 객체를 얻게 됨. 
# 이 경우 목적에 맞게 결과를 사용하려면 반드시 index[0]을 사용.
```
